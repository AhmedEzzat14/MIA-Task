```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb

WeatherDataset = pd.read_csv("Weather_Dataset.csv") # Read the weather file
WeatherDataset.dropna(inplace= True) # Handle the  missing  values
WeatherDataset.drop_duplicates(inplace= True) # Remove the duplicated values
WeatherDataset = WeatherDataset[WeatherDataset['Temperature (C)'] > -100] # Remove wrong values (Negative values)
WeatherDataset = WeatherDataset[WeatherDataset['Humidity'] > 0.3] # Remove wrong values (Negative values)
WeatherDataset = WeatherDataset.sort_values(by=['Formatted Date'],ascending=True) # Sorting the date
```

```python
x_axis = WeatherDataset['Formatted Date']
y_axis = WeatherDataset['Temperature (C)']
plt.plot(x_axis, y_axis)
```

```python
plt.hist(y_axis)
plt.show()
```

```python
x_axis2 = WeatherDataset['Temperature (C)']
y_axis2 = WeatherDataset['Humidity']
plt.scatter(x_axis2, y_axis2)
plt.show()
```

```python
WeatherDataset = WeatherDataset.select_dtypes(include=['float64','int64'])
sb.heatmap(WeatherDataset.corr())
plt.show 
```
