```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing

# Read data into weatherData variable
weatherHistory = pd.read_csv("weatherHistory.csv") 
weatherHistory.describe() # Gives basic statistics about our data
```

```python
WeatherDetails = ["Temperature (C)","Apparent Temperature (C)","Wind Speed (km/h)", "Wind Bearing (degrees)","Visibility (km)","Pressure (millibars)"]
x = weatherHistory[WeatherDetails]
y = weatherHistory.Humidity
# Scale data for more accurate
x_scaled  = preprocessing.scale(x)
# Polynomial Features enable us to capture non-linear relationships
polynomial = PolynomialFeatures(1)
x_final = polynomial.fit_transform(x_scaled)
# Split data into training and test data
# Split data into training and test data
x_Train, x_Test, y_Train, y_Test = train_test_split(x_final, y, test_size=0.10, random_state=42)
# Regularization parameter set to 0.5
Regression = linear_model.Ridge(alpha = 0.5) 
# Fit Model to Data
Regression.fit(x_Train, y_Train)
# Produce a set of predicted Humidity values from our test data
y_pred = Regression.predict(x_Test)
```
